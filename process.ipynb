```python
import pandas as pd
import numpy as np
from unidecode import unidecode
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
sort_names = [
        "Selection sort", "Insertion sort", "Bubble sort",
        "Shaker sort", "Shell sort",
        "Heap sort", "Merge sort", "Quick sort",
        "Counting sort", "Radix sort", "Flash sort"
    ]
data_order = [
        "Randomized",
        "Sorted",
        "Reversed",
        "Nearly sorted"
    ]
input_sizes = [10000, 30000, 50000, 100000, 300000, 500000]
```

```python
sort_data = []
for i in range(11):
    myFile = "output/output " + sort_names[i] + ".csv"
    df = pd.read_csv(myFile)
    sort_data.append(df)
sort_data[0]
```

```python
def generate_table(_type, sort_data):
    running_time_filtered = []
    table = [[],[]]
    for i in range(11):
        df = sort_data[i]
        _type_data_order = df[df["Data order"] == _type]
        for j in range(2):
            entry = {}
            entry["Name algorithm"] = sort_names[i]
            for _size in input_sizes[3*j:3*(j+1)]:
                raw_data = _type_data_order[_type_data_order["Data size"] == _size]
                entry[str(_size)+ '_1'] = raw_data["Running time"].values[0]
                entry[str(_size)+ '_2'] = raw_data["Comparisons"].values[0]
            table[j].append(entry)

    for i in range(2):
        running_time_filtered.append(pd.DataFrame(table[i]))

    return running_time_filtered
```

```python
for _type in data_order:
    running_time = generate_table(_type,sort_data)
    for i in range(2):
        # Convert the DataFrame to a LaTeX table
        latex_table = running_time[i].style.hide(axis="index").to_latex()
        # Split the LaTeX table into lines
        lines = latex_table.split('\n')

        # Insert \hline after each row (except the first and last)
        for j in range(0, len(lines) - 3, 1):
            lines[j] += ' \n\\hline'

        # Recreate the LaTeX table by joining the lines
        latex_table = '\n'.join(lines)
        with open('table latex/output_table_' + _type + '_' +str(i) + '.tex', 'w') as file:
            file.write(latex_table)
```
